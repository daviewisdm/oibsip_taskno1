```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Dataset Loading
df = pd.read_csv("AB_NYC_2019.csv")
df
```

```python
#Display first few rows
df.head()
```

```python
# Dataset Rows & Columns count
# Checking number of rows and columns of the dataset using shape
print("Number of rows are: ",df.shape[0])
print("Number of columns are: ",df.shape[1])
```

```python
# Dataset Info
# Checking information about the dataset using info
df.info()
```

```python
# Dataset Duplicate Value Count
dup = df.duplicated().sum()
print(f'number of duplicated rows are {dup}')
```

```python
# Missing Values/Null Values Count
df.isnull().sum()
```

```python
#Filling Missing Values
df['reviews_per_month'].fillna(0)
df['last_review'].fillna(0)
df['last_review_month'].fillna(0)
df['last_review_year'].fillna(0)
```

```python
# Missing Values/Null Values Count
df.isnull().sum()
```

```python
#Re-check for missing values
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
# Converting 'last_review' to datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
# Fill missing 'last_review' with the median of the available dates
median_last_review = df['last_review'].median()
df['last_review'].fillna(median_last_review)
```

```python
# Drop rows with missing 'name' or 'host_name'
df.dropna(subset=['name', 'host_name'], inplace=True)
```

```python
# Extract year, month, day from 'last_review'
df['last_review_year'] = df['last_review'].dt.year
df['last_review_month'] = df['last_review'].dt.month
df['last_review_day'] = df['last_review'].dt.day
```

```python
# Verify no missing values remain in 'name' or 'host_name'
missing_values_after = df.isnull().sum()
print("Missing Values After Cleaning:\n", missing_values_after)
```

```python
#Verifying the Data types
print(df.dtypes)
```

```python
#Descriptive Statistics
desctriptive_stats = df.describe()
print("Descriptive statistics:/n", desctriptive_stats)
```

### Visualisation

```python
# Plot distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot number of listings per neighborhood
plt.figure(figsize=(12, 8))
sns.countplot(y='neighbourhood_group', data=df, order=df['neighbourhood_group'].value_counts().index)
plt.title('Number of Listings per Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

```python
# Scatter plot of price vs. number of reviews
plt.figure(figsize=(10, 6))
plt.scatter(df['number_of_reviews'], df['price'], alpha=0.5)
plt.title('Price vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all listing names into one long string
text = " ".join(name for name in df.name)

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot the word cloud
plt.figure(figsize=(10, 5), facecolor=None)
plt.title('Most Used Words in Listing Names', fontsize=15, pad=20)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)

# Display the chart
plt.show()
```

```python
# Exclude non-numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=[np.number])

# Correlation Heatmap
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Outliner Detection

```python
#identify potential outliner in the 'Price' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Boxplot of Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Save cleaned dataset to a new CSV file
df.to_csv("AB_NYC_2019_Cleaned.csv", index=False)
```
